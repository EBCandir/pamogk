```python
from pathway_reader import kgml_pathway_reader
```

```python
all_pathways = kgml_pathway_reader.get_all_pathways()
pw_sizes = []
for pathway_id in all_pathways:
    entries, relations = kgml_pathway_reader.get_pathway_KGML(pathway_id)
    num_ent = len(entries.keys())
    print(pathway_id, num_ent, len(relations), '\n')
    if num_ent > 3000:
        print('Too big:', pathway_id)
    else:
        pw_sizes.append(num_ent)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

data = [go.Histogram(x=pw_sizes)]

layout = dict(title = 'KEGG Pathway Node Count Histogram',
              yaxis = dict(title='Bin Count', zeroline = False),
              xaxis = dict(title='Bin Sizes', zeroline = False)
             )

fig = dict(data=data, layout=layout)

py.iplot(fig, filename='Pathway Node Size Histogram')
```
